### yelp.com - Search for falafel in New York

```python
url = 'https://www.yelp.com/search?find_desc=falafel&find_loc=New+York%2C+NY&ns=1'
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
response = requests.get(url)
```

```python
response
```

```python
response.text[:100]
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup.find('span')
```

```python
soup.find_all('span')
```

```python
soup.find('a', {'data-analytics-label': 'biz-name'}).text
```

```python
rests = soup.find_all('a', {'data-analytics-label': 'biz-name'})
```

```python
for rest in rests:
    print(rest.text)
```
